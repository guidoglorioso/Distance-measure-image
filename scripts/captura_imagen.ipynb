# Captura de imagenes 
En este notebook se muestra un ejemplo de como obtener una imagen en un celular mediante la API de droidcam.
IMPORTANTE: 
* Descargar la aplicacion para android
* Abrirla y tenerla en segundo plano
* Configurar la IP segun la que haya sido asignada en la app

NOTA: si queres conseguir resoluciones de la version pro (ej 1280x720) podes ver una publicidad en la app del celular que te habilita durante 1 hora para usar esta resolucion.

```python
# Aseguro que cuando haya un cambio en la libreria se actualice

%load_ext autoreload
%autoreload 2
```

```python
from capture import IP_CAMERA 
import matplotlib.pyplot as plt
import time 

cam = IP_CAMERA("192.168.0.120",resolution=IP_CAMERA.SIZE1920x1080)
cam.init_camera()
cam.cmdSender(cam.toggleLED)
cam.cmdSender(cam.autoFocus)
time.sleep(1)
cam.set_rotation_angle(angle =360)
image = cam.get_img()

if image is not None:
    plt.imshow(image)
    
cam.disconnect()
```
